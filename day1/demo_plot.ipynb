```python
import numpy as np
import matplotlib.pyplot as plt
```

## Demos

### scatter() and plot()

```python
x = [0,2,5,4]
y = [1,3,2,1]
plt.scatter(x,y)
plt.xlabel('x')
plt.xlabel('y')
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.xlabel('y')
```

### Straight line

```python
x = np.linspace(-1,1,100)
print(x)

y2 = 2*x + 0.5
plt.plot(x, y2)
plt.xlabel('x')
plt.xlabel('y')
```
