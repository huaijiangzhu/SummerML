## Calculate the mean, variance, covariance

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-1,1,100)
y = x**2


xbar = np.mean(x)
xvar = np.var(x)

ybar = np.mean(y)
yvar = np.var(y)

covxy = np.mean( (x-xbar)*(y-ybar) )
```

import the data from the github

```python
url = 'https://raw.githubusercontent.com/huaijiangzhu/SummerML/master/day2/lab_stats.csv'
df = pd.read_csv(url)

df.head(5)
```

The data is organized into **cases**. Each case represents a different relationship between the variables x and y. We can choose the samples from a single case by using *logical indexing*.

```python
x = df['x'].values
y = df['y'].values
case = df['case'].values

i=7
log_ind = (case==i)

x0 = x[log_ind]
y0 = y[log_ind]

plt.plot(x0,y0,'ob',markeredgecolor='black')
plt.xlim([-2,2])
plt.ylim([-2,2])
plt.xlabel('x')
plt.ylabel('y');
plt.title('Case ' + str(i));
```
