```python
import numpy as np
import matplotlib.pyplot as plt
```

## Demos

### scatter() and plot()

```python
x1 = [0,2,5,4]
y1 = [1,3,2,1]
plt.scatter(x1,y1)
plt.xlabel('x')
plt.xlabel('y')
```

```python
plt.plot(x1,y1)
plt.xlabel('x')
plt.xlabel('y')
```

### Straight line

```python
x = np.linspace(-1,1,100)
print(x)

y2 = 2*x + 0.5
plt.plot(x, y2)
plt.xlabel('x')
plt.xlabel('y')
```

### Your task: 
Plot:
- sine wave
- polynomials
- exponential

```python
y3 = np.sin(2*np.pi*x)
plt.plot(x, y3)
```

```python
y4 = x**3 + 2
plt.plot(x, y4)
```

```python
y5 = np.exp(-2*x)
plt.plot(x, y5)
```

```python
## All-in-one plot


plt.figure()  #figsize = (8,8)
plt.plot(x, y2)  # c = 'green'  change color
plt.plot(x, y3)
plt.plot(x, y4)
plt.plot(x, y5)

plt.plot(x1, y1, 'o')
## After add scatter points, need to change x to (-1,5) to get better plot
plt.ylim([-2, 4])
plt.xlim([-1, 5])
plt.grid()
plt.legend(['Linear','Sine','Poly','Expo','Scatter'])
```
